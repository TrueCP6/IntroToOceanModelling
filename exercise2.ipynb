# Exercise 2
We plot the analytical solution to a system of ODEs:
$$
P(t) = \frac{P_0(P_0 + N_0)}{P_0 + N_0 \exp (-k(P_0 + N_0)t)}
$$
$$
N(t) = P_0 + N_0 - P(t)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

P_0 = 0.1
N_0 = 5
k = 0.1
t = np.linspace(0, 20)

def P_fun(t, k):
    return P_0*(P_0 + N_0)/(P_0 + N_0*np.exp(-k*(P_0 + N_0)*t))
def N_fun(t, k):
    return P_0 + N_0 - P_fun(t,k)

P = P_fun(t, k)
N = N_fun(t, k)

plt.plot(t, P, label="P")
plt.plot(t, N, label="N")
plt.title('Concentration of P and N over time')
plt.xlabel('Time (d)')
plt.ylabel('Concentration (mmol N m^-3)')
```

## Question 1
The curves are identical in shape because the one is a linear scaling of the other, i.e $N(t) = A*P(t)+B$. This is because in the original system of ODEs $dP/dt = - dN/dt$, meaning the gradients of each curve will simply be the negative inverse of each other.

## Question 2
We change the value of the update parameter:

```python
k = 0.2
P = P_fun(t, k)
N = N_fun(t, k)

plt.plot(t, P, label="P")
plt.plot(t, N, label="N")
plt.title('Concentration of P and N over time')
plt.xlabel('Time (d)')
plt.ylabel('Concentration (mmol N m^-3)')
```

We see that increasing $k$ decreases the time taken to reach equilibrium. Decreasing $k$ has the opposite effect.
