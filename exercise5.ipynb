# Exercise 5
We check that the implicit Euler backward scheme is unconditionally stable:
$$
C^{n+1} = \frac{C^n}{1+k \Delta t}
$$

We know that $k,\Delta t> 0$, and hence $1+k\Delta t > 1$. This means that $r = \frac{1}{1 + t \Delta} < 1$. Since every $C^{n+1} < C^n$, and the nature of the series is geometric, $C^n \to 0$ as $n \to \infty$ and hence the scheme is stable.
