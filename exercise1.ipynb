To answer the questions, first we must import the data and plot it.


```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('data/flynn94.csv', delimiter=',', dtype=float, filling_values=np.nan, names=True, encoding='utf-8-sig')
t = data['time']

plt.scatter(t, data['cells'])
plt.title('Cell Count vs Time')
plt.xlabel('Time')
plt.ylabel('Cells')

plt.figure()
plt.scatter(t, data['biovolume'])
plt.title('Biovolume vs Time')
plt.xlabel('Time')
plt.ylabel('Biovolume')

plt.figure()
plt.scatter(t, data['ammonium'])
plt.title('Ammonium vs Time')
plt.xlabel('Time')
plt.ylabel('Ammonium')

plt.figure()
plt.scatter(t, data['cellC'])
plt.title('Cell Concentration vs Time')
plt.xlabel('Time')
plt.ylabel('Cell Concentration')

plt.figure()
plt.scatter(t, data['chla'])
plt.title('Chlorophyll a vs Time')
plt.xlabel('Time')
plt.ylabel('Chlorophyll a')
```
