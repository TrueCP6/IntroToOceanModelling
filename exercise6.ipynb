We implement the backward Euler scheme for the equation $\frac{dC}{dt} = -kC$, and plot the results for multiple k.

```python
import numpy as np
import matplotlib.pyplot as plt

def backward_euler(C0, k, dt, T):
    n_steps = int(T / dt)
    t = np.linspace(0, T, n_steps + 1)
    C = np.zeros(n_steps + 1)
    C[0] = C0

    for i in range(1, n_steps + 1):
        C[i] = C[i-1] / (1 + k * dt)

    return t, C

# parameters
C0 = 1.0 # initial value
dt = 0.1 # time step
T = 10.0 # max time
k_values = [0.1, 0.5, 1.0, 2.0]  # k values to plot


plt.figure()
for k in k_values:
    t, C = backward_euler(C0, k, dt, T)
    plt.plot(t, C, label=f'k = {k}', linewidth=2)

plt.title('Backward Euler Scheme for dC/dt = -kC')
plt.xlabel('Time (t)')
plt.ylabel('C(t)')
plt.legend()
```

We see that increasing $k$ simply increases the rate with which the solution approaches zero.
