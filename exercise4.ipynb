# Exercise 4
We compute the Taylor series about $x = x_0$ for the following functions using WolframAlpha:

1. $f(x)= \frac{x}{x+1}$:
$$
f(x) = \frac{x_0}{x_0 + 1} + \frac{x - x_0}{(x_0+1)^2} + O\left((x-x_0)^3\right)
$$

2. $g(x) = \frac{2x^2}{x^4} = 2x^2$:
$$
g(x) = 2x_0 ^2 + 4 x_0 (x - x_0) + 2(x-x_0)^2
$$

3. $h(x) = \ln(x^2)$:
$$
h(x) = \ln(x_0 ^2) + \frac{2(x-x_0)}{x_0} - \frac{(x-x_0) ^2}{x_0 ^2} + O((x-x_0)^3)
$$
