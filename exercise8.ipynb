# Exercise 8
We derive the backward Taylor expansion for $U(x)$
$$
U(x) = U(x_0) + U'(x_0) (x-x_0) + O((x-x_0)^2)
$$
We set $x \to x - \Delta x$ and $x_0 \to x$, as $x-x_0= \Delta x$
$$
U(x - \Delta x) = U(x) - U'(x) \Delta x + O(\Delta x^2)
$$

Combining these equations and rearranging yields
$$
\Delta x U'(x) = U(x) - U(x - \Delta x) + O(\Delta x^2)
$$
which divided by $\Delta x$ is
$$
U'(x) = \frac{U(x) - U(x - \Delta x)}{\Delta x} + O(\Delta x)
$$
which is the same as
$$
\frac{\partial U}{\partial x}\vert^n_i = \frac{U^n_i - U^n_{i-1}}{\Delta x} + O(\Delta x)
$$
