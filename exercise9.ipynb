We write the second order central difference approximation to the second derivative.
We define $t^n = n \Delta t$ and $x_i = i \Delta x$

$$
\frac{\partial ^2 U}{\partial x^2} = \frac{U_{i-1}^n - 2U_i^n + U_{i+1}^n}{\Delta x^2} + O(\Delta x^2)
$$

We write equation (33) in terms of the discrete velocity field $U_{i,j,k}^n$:
$$
\frac{\partial U}{\partial z} = \frac{U_{i,j,k-1}^n -2 U_{i,j,k}^n + U_{i,j,k+1}^n}{2 \Delta z} + O(\Delta z^2)
$$
