# Exercise 13
We will discretise the equations for the Ekman velocity:
$$\frac{\partial u_E}{\partial t} - f v_E = A_v \frac{\partial^2 u_E}{\partial z^2}$$
$$\frac{\partial v_E}{\partial t} + f u_E = A_v \frac{\partial^2 v_E}{\partial z^2}$$

## Question 1
We can discretise these equations using a forward Euler scheme and a second order central difference stencil for the second derivative. The $E$ in the subscript will be omitted for readability.

$$\frac{u^{n+1}_k - u^n_k}{\Delta t} - f v^n_k = A_v \frac{u^n_{k-1} -2 u^n_{k} + u^n_{k+1}}{\Delta z^2}$$
$$\frac{v^{n+1}_k - v^n_k}{\Delta t} + f u^n_k = A_v \frac{v^n_{k-1} -2 v^n_{k} + v^n_{k+1}}{\Delta z^2}$$

We can rewrite this discretisation to solve for $u^{n+1}$ and $v^{n+1}$:
$$u^{n+1}_k = u^n_k + \Delta t A_v \frac{u^n_{k-1} -2 u^n_{k} + u^n_{k+1}}{\Delta z^2} + \Delta t f v^n_k$$
$$v^{n+1}_k = v^n_k + \Delta t A_v \frac{v^n_{k-1} -2 v^n_{k} + v^n_{k+1}}{\Delta z^2} - \Delta t f u^n_k$$

## Question 2
To determine the Courant number, we take the coefficients of the above equations. Since $f \ll A_v$, we can neglect the effect of the source term, and hence the Courant number for the Ekman velocity equations is $C=\frac{A_v \Delta t}{\Delta z^2}$.

## Question 3
If we assume $A_v = 10^{-4} \, \text{m}^2/\text{sec}$, we need $C \leq \frac{1}{2}$ for the stability of the . This is equivalent to the inequality $\Delta t < \frac{\Delta z^2}{2 A_v} = 10^4 \Delta z^2$. For example, for $\Delta z = 2 \,\text{m}$, we have the constraint on the diffusion term that $\Delta t < 2 \times 10^4$. We also have the constraint on the source term, where $f \Delta t \ll 1$. This means that in practice for $\Delta z =2$ we would need a much smaller timestep on the order of $\Delta t \approx 160 $ sec.
